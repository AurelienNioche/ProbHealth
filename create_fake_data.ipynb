```python
import pandas as pd
import numpy as np
```

```python
n_obs = 1000

theta1 = np.array([[1., 2., 4.], [-3., -2., 4.]])
theta2 = np.array([5., -3., 4.])
cst = -3.

m = np.random.random(size=(n_obs, 3))

x = np.random.random(size=n_obs)

noise = np.random.normal(0, 0.001, size=(n_obs, 3))
noise2 = np.random.normal(0, 0.002, size=n_obs)

m = np.zeros((n_obs, 3))
for j in range(3):
    m[:, j] = theta1[0, j] + x * theta1[1, j] + noise[:, j]

y = np.zeros(n_obs) + cst + noise2
for j in range(3):
    y[:] += m[:, j] * theta2[j] 
```
